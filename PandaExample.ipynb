```python
import pandas as pd

# 1. Create a dictionary of data
data = {
    'Name': ['John', 'Alice', 'Bob', 'David'],
    'Age': [25, 30, 35, 28],
    'City': ['New York', 'London', 'Paris', 'Tokyo'],
    'Salary': [60000, 75000, 80000, 62000]
}

# 2. Create a pandas DataFrame from the dictionary
df = pd.DataFrame(data)

# 3. Print the original DataFrame
print("Original DataFrame:")
print(df)
print("-" * 20)

# 4. Filter the DataFrame
# Select only the 'Name' and 'City' columns for people older than 28
# We use boolean indexing for the rows and a list for column selection
filtered_df = df.loc[df['Age'] > 28, ['Name', 'City', 'Age']]

# 5. Print the filtered DataFrame
print("\nFiltered DataFrame (People older than 28):")
print(filtered_df)
```
